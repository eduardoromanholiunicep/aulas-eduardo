## Exemplo Resolvido 1: Tabuada

**O problema:** Crie um programa que peça um número ao usuário e imprima a tabuada desse número de 1 a 10 usando o `for`.

```python
numero = int(input("Você quer a tabuada de qual número? "))

print(f"\n--- Tabuada do {numero} ---")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i:2d} = {resultado:2d}")
```

*Dica: O formato `{:2d}` dentro da f-string diz para o Python reservar 2 espaços na hora de imprimir aquele número. Isso ajuda a deixar os números alinhados bonitinhos na tela!*

## Exemplo Resolvido 2: Somatória com While

**O problema:** Peça números ao usuário repetidamente. O programa deve somar todos os números digitados e só deve parar (e mostrar o total) quando o usuário digitar o **número ZERO**.

```python
soma = 0
while True:
    valor = float(input("Digite um valor para somar (ou 0 para parar): "))

    if valor == 0:
        break # Sai do loop quando o usuário digitar zero

    soma = soma + valor

print(f"\nA soma total foi de: {soma}")
```

### Desafio 2: Validação de Nota
Faça um programa que peça para o usuário digitar uma nota entre 0 e 10.
Sempre que o usuário digitar um valor inválido (menor que 0 ou maior que 10), o programa deve imprimir "Nota Inválida" e pedir para ele digitar a nota novamente. Só deve aceitar e encerrar o programa quando a nota for válida.

### Desafio 3.
Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

```python
populacao_a = 80000
taxa_crescimento_a = 0.03
populacao_b = 200000
taxa_crescimento_b = 0.015
anos = 0

while populacao_a < populacao_b:
    populacao_a += populacao_a * taxa_crescimento_a
    populacao_b += populacao_b * taxa_crescimento_b
    anos += 1

print(f"Serão necessários {anos} anos para que a população de A ultrapasse ou iguale a população de B.")
```

### Desafio 10.
Faça um programa que receba dois números inteiros e gere os números inteiros que estão no intervalo compreendido por eles.

```python
inicio = int(input("Digite o primeiro número inteiro: "))
fim = int(input("Digite o segundo número inteiro: "))

# Ajustando a ordem caso o primeiro seja maior que o segundo
if inicio > fim:
    inicio, fim = fim, inicio

print(f"\nNúmeros no intervalo entre {inicio} e {fim}:")
for i in range(inicio + 1, fim):
    print(i, end=" ")
```

### Desafio 12.
Desenvolva um gerador de tabuada, capaz de gerar a tabuada de qualquer número inteiro entre 1 a 10. O usuário deve informar de qual numero ele deseja ver a tabuada. A saída deve ser conforme o exemplo abaixo:

Tabuada de 5:

5 X 1 = 5
5 X 2 = 10
...

5 X 10 = 50

```python
while True:
    try:
        numero_str = input("Informe um número inteiro entre 1 e 10 para ver a tabuada: ")
        numero = int(numero_str)

        if 1 <= numero <= 10:
            print(f"\nTabuada de {numero}:\n")
            for i in range(1, 11):
                resultado = numero * i
                print(f"{numero} X {i:2d} = {resultado:2d}")

            continuar = input("\nDeseja gerar outra tabuada? (s/n): ").lower()
            if continuar != 's':
                break
        else:
            print("Número inválido! Por favor, digite um número entre 1 e 10.")
    except ValueError:
        print("Entrada inválida! Por favor, digite um número inteiro.")
```

### Desafio 13.
Faça um programa que peça dois números, base e expoente, calcule e mostre o primeiro número elevado ao segundo número. Não utilize a função de potência da linguagem.

```python
base = int(input("Digite o número base: "))
expoente = int(input("Digite o expoente (número inteiro não negativo): "))

resultado = 1
# Se o expoente for 0, qualquer número elevado a 0 é 1 (exceto 0^0 que é indefinido, mas aqui consideramos 1)
if expoente == 0:
    resultado = 1
elif expoente > 0:
    for _ in range(expoente):
        resultado *= base  # Multiplica o resultado pela base 'expoente' vezes
else:
    # Para expoentes negativos, o cálculo é 1 / (base ^ |expoente|)
    # Por simplicidade, vamos focar em expoentes não negativos conforme o problema geralmente implica.
    # Se precisar de expoentes negativos, a lógica seria: 1 / (base * base * ...)
    print("Expoentes negativos não são suportados neste exemplo simples.")
    resultado = "Indefinido ou requer cálculo fracionário"

print(f"\n{base} elevado a {expoente} é: {resultado}")
```

### Desafio 14.
Faça um programa que peça 10 números inteiros, calcule e mostre a quantidade de números pares e a quantidade de números impares.
