## Exemplo Resolvido 1: Ano Bissexto

**O problema:** Crie um programa que peça um ano (ex: 2024) e informe se ele é Bissexto ou não.

*Regra do Ano Bissexto:*
Ele precisa ser múltiplo de 4 (`ano % 4 == 0`).
Mas se for fim de século (múltiplo de 100), ele **não** é bissexto, a menos que ele também seja múltiplo de 400.

```python
# Solução do Professor
ano = int(input("Digite o ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} É bissexto!")
else:
    print(f"O ano {ano} NÃO é bissexto.")
```

**O problema:** Peça dois números e um operador matemático (`+`, `-`, `*` ou `/`) ao usuário. Imprima o resultado da operação.

```python
num1 = float(input("Digite o primeiro número: "))
op = input("Digite a operação (+, -, *, /): ")
num2 = float(input("Digite o segundo número: "))

if op == "+":
    print(f"Resultado: {num1 + num2}")
elif op == "-":
    print(f"Resultado: {num1 - num2}")
elif op == "*":
    print(f"Resultado: {num1 * num2}")
elif op == "/":
    # Cuidados extras em programação: não se divide por zero!
    if num2 == 0:
        print("Erro: Divisão por zero.")
    else:
        print(f"Resultado: {num1 / num2}")
else:
    print("Operador inválido!")
```

### Exercício 9: Validação de Data (AND, OR, NOT)
Peça o dia, mês e ano. Verifique se a data é válida. Considere:
- Mês entre 1 e 12.
- Dia entre 1 e 31.
- Dias específicos para meses (30 dias para abril, junho, setembro, novembro; 29 em fevereiro de ano bissexto, 28 em fevereiro de ano não bissexto).
- Um ano é bissexto se for divisível por 4, mas não por 100, a menos que seja divisível por 400.
Imprima "Data Válida" ou "Data Inválida" e o motivo se possível.

```python
dia = int(input("Digite o dia: "))
mes = int(input("Digite o mês: "))
ano = int(input("Digite o ano: "))

data_valida = True
motivo_invalido = ""

# 1. Validar mês
if not (1 <= mes <= 12):
    data_valida = False
    motivo_invalido = "Mês inválido"
else:
    # 2. Validar dia com base no mês e ano
    if mes == 2: # Fevereiro
        # Verificar ano bissexto
        if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
            # Ano bissexto, fevereiro tem 29 dias
            if not (1 <= dia <= 29):
                data_valida = False
                motivo_invalido = "Dia inválido para Fevereiro (ano bissexto)"
        else:
            # Ano não bissexto, fevereiro tem 28 dias
            if not (1 <= dia <= 28):
                data_valida = False
                motivo_invalido = "Dia inválido para Fevereiro (ano não bissexto)"
    elif mes in [4, 6, 9, 11]: # Meses com 30 dias
        if not (1 <= dia <= 30):
            data_valida = False
            motivo_invalido = "Dia inválido para o mês (apenas 30 dias)"
    else: # Meses com 31 dias
        if not (1 <= dia <= 31):
            data_valida = False
            motivo_invalido = "Dia inválido para o mês (apenas 31 dias)"

if data_valida:
    print("Data Válida")
else:
    print(f"Data Inválida: {motivo_invalido}")
```

### Exercício 10: Prioridade de Processo (Múltiplas Condições Complexas)
Um sistema gerencia a prioridade de processos com base em vários critérios:
- **Tipo de Processo:** 'Crítico', 'Normal', 'Baixo'
- **Tempo de Espera:** Em minutos (inteiro)
- **Recursos Necessários:** 'Alto', 'Médio', 'Baixo'

Defina a prioridade final (1=Mais Alta, 5=Mais Baixa) usando as seguintes regras:
1. Se o tipo for 'Crítico': Prioridade 1.
2. Se o tempo de espera for maior que 60 minutos E o tipo for 'Normal': Prioridade 2.
3. Se os recursos necessários forem 'Alto' E o tipo NÃO for 'Crítico': Prioridade 3.
4. Se o tipo for 'Normal' E o tempo de espera for entre 30 e 60 minutos (inclusive): Prioridade 3.
5. Se o tipo for 'Baixo' OU os recursos necessários forem 'Baixo' (e nenhuma das anteriores se aplicar): Prioridade 5.
6. Caso contrário (nenhuma das regras acima se aplicar): Prioridade 4.

Peça ao usuário o tipo de processo, tempo de espera e recursos necessários, e imprima a prioridade final.

```python
tipo_processo = input("Digite o tipo de processo ('Crítico', 'Normal', 'Baixo'): ")
tempo_espera = int(input("Digite o tempo de espera em minutos: "))
recursos_necessarios = input("Digite os recursos necessários ('Alto', 'Médio', 'Baixo'): ")

prioridade = 4 # Valor padrão se nenhuma regra superior for aplicada

# Regra 1: Se o tipo for 'Crítico': Prioridade 1.
if tipo_processo.lower() == 'crítico':
    prioridade = 1
# Regra 2: Se o tempo de espera for maior que 60 minutos E o tipo for 'Normal': Prioridade 2.
elif tempo_espera > 60 and tipo_processo.lower() == 'normal':
    prioridade = 2
# Regra 3: Se os recursos necessários forem 'Alto' E o tipo NÃO for 'Crítico': Prioridade 3.
elif recursos_necessarios.lower() == 'alto' and tipo_processo.lower() != 'crítico':
    prioridade = 3
# Regra 4: Se o tipo for 'Normal' E o tempo de espera for entre 30 e 60 minutos (inclusive): Prioridade 3.
elif tipo_processo.lower() == 'normal' and 30 <= tempo_espera <= 60:
    prioridade = 3
# Regra 5: Se o tipo for 'Baixo' OU os recursos necessários forem 'Baixo' (e nenhuma das anteriores se aplicar): Prioridade 5.
elif tipo_processo.lower() == 'baixo' or recursos_necessarios.lower() == 'baixo':
    prioridade = 5
# Regra 6: Caso contrário (nenhuma das regras acima se aplicar): Prioridade 4. (já definido como padrão)

print(f"A prioridade final do processo é: {prioridade}")
```
